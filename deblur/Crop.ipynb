```python
from PIL import Image
import os

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
height = 128
width = 128
im = Image.open('./imgs/0801.png')
(img_width, img_height) = im.size
```

```python
boxMap = {}
k = 0
for i in range(0, img_height, height):
    for j in range(0, img_width, width):
        box = (j, i, j+width, i+height)
        boxMap['img' + str(k)] = box
        a = im.crop(box)
        a.save('./temp/{}.png'.format(k))
        k += 1
#         input(">img{}".format(k))

print(list(boxMap.keys()))
```
